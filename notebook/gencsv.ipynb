```python
import pandas as pd
import json
import xmltodict
pd.options.display.max_columns = 1000
import glob
```

```python
directories = glob.glob("../data/2008/2008/RCS-A*")
globarr1 = []
globarr2 = []
mydict = {}
for d in directories:
    print(d)
    with open(d, encoding='windows-1252') as fd:
        doc = xmltodict.parse(fd.read())
    bodaccdict = json.loads(json.dumps(doc))
    for i in range(len(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'])):
        mydict['originalFile'] = d.split("../data/2008/2008/")[1]
        mydict['parution'] = bodaccdict['RCS-A_IMMAT']['parution']
        mydict['dateParution'] = bodaccdict['RCS-A_IMMAT']['dateParution']
        mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
        mydict['data'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]
        globarr1.append(mydict)
        mydict = {}
        if(type(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']) is dict):
            if "personneMorale" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']:
                if(type(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personneMorale']) is dict):

                    if "numeroImmatriculation" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personneMorale']:
                        mydict['siren'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personneMorale']['numeroImmatriculation']['numeroIdentification']
                        mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
                        mydict['type'] = 'personneMorale'
                        globarr2.append(mydict)
                        mydict = {}
                else:
                    for j in range(len(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personneMorale'])):
                        if "numeroImmatriculation" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personneMorale'][j]:
                            mydict['siren'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personneMorale'][j]['numeroImmatriculation']['numeroIdentification']
                            mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
                            mydict['type'] = 'personneMorale'  
                            globarr2.append(mydict)
                            mydict = {}
                               
                        
            if "personnePhysique" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']:
                if(type(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personnePhysique']) is dict):
                    if "numeroImmatriculation" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personnePhysique']:
                        mydict['siren'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personnePhysique']['numeroImmatriculation']['numeroIdentification']
                        mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
                        mydict['type'] = 'personnePhysique'
                        globarr2.append(mydict)
                        mydict = {}
                else:
                    for j in range(len(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personnePhysique'])):
                        if "numeroImmatriculation" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personnePhysique'][j]:
                            mydict['siren'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']['personnePhysique'][j]['numeroImmatriculation']['numeroIdentification']
                            mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
                            mydict['type'] = 'personnePhysique'    
                            globarr2.append(mydict)
                            mydict = {}              

                
        if(type(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne']) is list):
            for k in range(len(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'])):
                if "personneMorale" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]:
                    if(type(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personneMorale']) is dict):
                        if "numeroImmatriculation" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personneMorale']:
                            mydict['siren'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personneMorale']['numeroImmatriculation']['numeroIdentification']
                            mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
                            mydict['type'] = 'personneMorale'
                            globarr2.append(mydict)
                            mydict = {}
                    else:
                        for j in range(len(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personneMorale'])):
                            if "numeroImmatriculation" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personneMorale'][j]:
                                mydict['siren'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personneMorale'][j]['numeroImmatriculation']['numeroIdentification']
                                mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
                                mydict['type'] = 'personneMorale'  
                                globarr2.append(mydict)
                                mydict = {}


                if "personnePhysique" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]:
                    if(type(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personnePhysique']) is dict):
                        if "numeroImmatriculation" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personnePhysique']:
                            mydict['siren'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personnePhysique']['numeroImmatriculation']['numeroIdentification']
                            mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
                            mydict['type'] = 'personnePhysique'
                            globarr2.append(mydict)
                            mydict = {}
                    else:
                        for j in range(len(bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personnePhysique'])):
                            if "numeroImmatriculation" in bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personnePhysique'][j]:
                                mydict['siren'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['personnes']['personne'][k]['personnePhysique'][j]['numeroImmatriculation']['numeroIdentification']
                                mydict['nojo'] = bodaccdict['RCS-A_IMMAT']['listeAvis']['avis'][i]['nojo']
                                mydict['type'] = 'personnePhysique'        
                                globarr2.append(mydict)
                                mydict = {}          
```
